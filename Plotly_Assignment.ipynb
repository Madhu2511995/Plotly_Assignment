```python
import plotly.graph_objects as go
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=titanic.age,y=titanic.fare,mode='markers'))
fig.show()
```

```python
Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
```

```python
tips=sns.load_dataset("tips")
```

```python
fig=go.Figure()
fig.add_trace(go.Box(x=tips.total_bill,y=tips.tip))
fig.show()
```

```python
import plotly.express as px
```

```python
fig = px.box(tips, x="day", y="total_bill", points="all")
# Update the layout for better visualization
fig.update_layout(
    title="Box Plot of Total Bill by Day",
    xaxis_title="Day",
    yaxis_title="Total Bill",
)
# Show the plot
fig.show()
```

```python
Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.
```

```python
fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker",
    title="Histogram of Total Bill by Sex",
    labels={"sex": "Sex", "total_bill": "Total Bill"},
)

# Show the plot
fig.show()
```

```python
Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.
```

```python
iris = sns.load_dataset("iris")

# Create the scatter matrix plot using Plotly Express
fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset",

)

# Show the plot
fig.show()
```

```python
Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
import plotly.figure_factory as ff
import numpy as np

x1 = np.random.randn(100)
x2 = np.random.randn(100)
x3 = np.random.randn(100) + 1

hist_data = [x1, x2, x3]

group_labels = ['Group 1', 'Group 2', 'Group 3']
colors = ['#333F44', '#37AA9C', '#94F3E4']

# Create distplot with curve_type set to 'normal'
fig = ff.create_distplot(hist_data, group_labels, show_hist=False, colors=colors)

# Add title
fig.update_layout(title_text='Curve and Rug Plot')
fig.show()
```
